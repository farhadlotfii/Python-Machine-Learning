# Simple vs. Miltiple Linear Regression
## Simple Linear Regression (SLR)

```python
X = data[['RM']]
Y = data[['PRICE']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(x_train, y_train)
```

```python
slr.intercept_
# b0
```

```python
slr.coef_
# b
```

```python
coefficients = pd.DataFrame([x_train.columns, slr.coef_]).T
coefficients = coefficients.rename(columns={0:'Attribute', 1:'coefficients'})
coefficients
```

## SLR Evaluation

```python
from sklearn import metrics
```

```python
y_pred = slr.predict(x_test)
```

```python
print('R^2 squared : ', metrics.r2_score(y_test, y_pred))
print('MAE : ', metrics.mean_absolute_error(y_test, y_pred))
print('MSE : ', metrics.mean_squared_error(y_test, y_pred))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('prices')
plt.ylabel('predicted price')
plt.title('Prices vs. Predicted Prices')
```

```python
sns.displot(y_test - y_pred)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
```

## Multiple Linear Regression (MLR)

```python
X = data.drop(['PRICE'], axis=1)
Y = data['PRICE']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
mlr.intercept_
# b0
```

```python
mlr.coef_
# b
```

```python
coefficients = pd.DataFrame([x_train.columns, mlr.coef_]).T
coefficients = coefficients.rename(columns={0:'Attribute', 1:'coefficients'})
coefficients
```

```python
y_pred = mlr.predict(x_test)
```

```python
from sklearn import metrics

print('R^2 squared : ', metrics.r2_score(y_test, y_pred))
print('MAE : ', metrics.mean_absolute_error(y_test, y_pred))
print('MSE : ', metrics.mean_squared_error(y_test, y_pred))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('prices')
plt.ylabel('predicted price')
plt.title('Prices vs. Predicted Prices')
```

```python
sns.displot(y_test - y_pred)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# create a polynomial regression model for given degree
poly_features = PolynomialFeatures(degree=2)

x_train_quadratic = poly_features.fit_transform(x_train)
x_test_quadratic = poly_features.fit_transform(x_test)

quadratic = LinearRegression()
quadratic.fit(x_train_quadratic, y_train)


y_pred_train = quadratic.predict(x_train_quadratic)
y_pred_test = quadratic.predict(x_test_quadratic)
```

```python
metrics.r2_score(y_test, y_pred_test)
```

```python
metrics.r2_score(y_train, y_pred_train)
```

```python
plt.scatter(y_test, y_pred_test)
plt.xlabel('prices')
plt.ylabel('predicted price')
plt.title('Prices vs. Predicted Prices')
```

```python
sns.displot(y_test - y_pred_test)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
```

```python
poly_features = PolynomialFeatures(degree=3)

x_train_quadratic = poly_features.fit_transform(x_train)
x_test_quadratic = poly_features.fit_transform(x_test)

quadratic = LinearRegression()
quadratic.fit(x_train_quadratic, y_train)


y_pred_train = quadratic.predict(x_train_quadratic)
y_pred_test = quadratic.predict(x_test_quadratic)
```

```python
metrics.r2_score(y_test, y_pred_test)
```

```python
metrics.r2_score(y_train, y_pred_train)

# overfit
```
